```python
import pandas as pd
import glob
import os
```

## Programming.in.th

```python
csv_dir = './programmin/'
csv_files = glob.glob(os.path.join(csv_dir, '*.csv'))

df = pd.concat([pd.read_csv(file) for file in csv_files], ignore_index=True)
df
```

```python
# Remove all non-cpp files
df = df[df['Language']=="C++"]
df
```

```python
# Remove all 0 point
df = df[df['Score']!='0 points']
df
```

```python
# Remove 0000 , 0001 , 0002 ... 0005 , codecube_001 codecube_002
df = df[~df['ProblemURL'].isin([
    "0000", "0001", "0002", "0003", "0004", "0005",
    "codecube_001", "codecube_002"
])]

df['ProblemURL'].value_counts()
```

```python
df['Score'] = df['Score'].str.extract(r'(\d+)').astype(int)
```

```python
# Check TOI Problems
df[df['ProblemURL'].str.contains('toi', case=False, na=False)]
```

Label with writer,problem_id,submission_id,website,source (human, - , - , https://programming.in.th/ , -)

```python
df = df.drop_duplicates(subset='Source',keep='last')
```

```python
df = df.sort_values('Score', ascending=False)
df = df.drop_duplicates(subset=['User', 'ProblemURL'], keep='first')
```

```python
df
```

```python
df['User'].value_counts()[:15]
```

```python
cleaned_df = pd.DataFrame()

cleaned_df['problem_id'] = df['ProblemURL']
cleaned_df['submission_id'] = df['SubmissionID']
cleaned_df['source'] = df['Source']
cleaned_df['writer'] = "human"
cleaned_df['website'] = "https://programming.in.th"
cleaned_df = cleaned_df[['writer','problem_id','submission_id','website','source']]
cleaned_df
```

```python
cleaned_df.to_csv('./cleaned/programmingin1.csv',index=False)
```

## Grader Chan

```python
df = pd.read_csv('./graderchan_source.csv')
df
```

```python
df = df[(df['Language'] == "Language: cpp")]
df
```

```python
df = df[df['Score'] != "Score: 0"]
df
```

```python
df['Score'] = df['Score'].str.extract(r'Score:\s*(\d+)').astype(int)
```

```python
df = df.drop_duplicates(subset='Source',keep='last')
```

```python
df = df.sort_values('Score', ascending=False)
df = df.drop_duplicates(subset=['User', 'ProblemURL'], keep='first')
```

```python
df['User'].value_counts()[:20]
```

```python
cleaned_df = pd.DataFrame()

cleaned_df['problem_id'] = df['ProblemURL']
cleaned_df['submission_id'] = df['SubmissionID']
cleaned_df['source'] = df['Source']
cleaned_df['writer'] = "human"
cleaned_df['website'] = "https://firefly.gchan.moe"
cleaned_df = cleaned_df[['writer','problem_id','submission_id','website','source']]
cleaned_df
```

```python
cleaned_df.to_csv('./cleaned/graderchan1.csv',index=False)
```
