```python
import json
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Source Codes (Human)

```python
source_code_path = "./../Datasets/Humans/Cleaned/"
df = pd.DataFrame()

HumanDFs = []

# for p in os.listdir(source_code_path):
#     if p.endswith(".csv"):
#         df = pd.concat([df, pd.read_csv(source_code_path + p)])

for p in os.listdir(source_code_path):
    if p.endswith(".csv"):
        HumanDFs.append(pd.read_csv(source_code_path + p))
```

```python
CombinedHumanDF = pd.DataFrame()
for df in HumanDFs:
    CombinedHumanDF = pd.concat([CombinedHumanDF,df])
CombinedHumanDF
```

```python
CombinedHumanDF['submission_id'] = pd.to_numeric(CombinedHumanDF['submission_id'], errors='coerce').astype('Int64')
CombinedHumanDF
```

## Load Source Codes (AI)

```python
source_code_path = "./../datasets/AI-Generated/Cleaned/"
df = pd.DataFrame()

AIDFs = []

for p in os.listdir(source_code_path):
    if p.endswith(".csv"):
        AIDFs.append(pd.read_csv(source_code_path + p))
    
AIDFs[0]
```

```python
CombinedAIDF = pd.DataFrame()
for df in AIDFs:
    CombinedAIDF = pd.concat([CombinedAIDF,df])
CombinedAIDF
```

```python
CombinedAIDF['submission_id'] = pd.to_numeric(CombinedAIDF['submission_id'], errors='coerce').astype('Int64')
CombinedAIDF
```

## Clean NAN

```python
CombinedHumanDF = CombinedHumanDF[~CombinedHumanDF['source'].isna()]
CombinedAIDF = CombinedAIDF[~CombinedAIDF['source'].isna()]
```

## Clean Duplicate

```python
CombinedAIDF = CombinedAIDF[~CombinedAIDF['source'].duplicated(keep='first')]
CombinedHumanDF = CombinedHumanDF[~CombinedHumanDF['source'].duplicated(keep='first')]
```

```python
CombinedAIDF.to_csv('./combined/AI.csv')
CombinedHumanDF.to_csv('./combined/Human.csv')
```
